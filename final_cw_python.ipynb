# Task 3

```python
M = np.array([[3,2,-1],[2,-1,4],[4,-2,3]])
X1 = np.array([['x'],['y'],['z']]) 
C1 = np.array([[25],[19],[18]])
print("M:\n",M, "\nX=\n",X1,"\nC:\n",C1)
```

# Task4

```python
#Matrix with the coefficints(M) is multiplied by matrix with the variables(X1)
#this matrix product is equal to the matrix of column of constants(C1)
#M matrix has order 3*3 and X1 has order 3*1
#as the no. of columns of matrix M is equal to no. of rows of matrix X1, These matrices are compatible for multiplication
#the order of matrix multiplication will be 3*1
#to solve these matrices we multiply m-inverse both the sides
#Mulplication of inverse matrix with itself is a Identity matrix.
#so the final equation becomes X1= Inv_M * C1
```

# Task5

```python
#Inverse of matrix M is calculated as follows
Inv_M =  np.linalg.inv(M)
```

```python
Inv_M
```

```python
#matrix with the variables(X1) is equal to multiplication of inverse of Matrix with the coefficints(M) with matrix of constants(C1)
#X1= Inv_M * C1
X1=np.matmul(Inv_M,C1)
```

```python
X1
```

```python
print("Values of x,y,z are:\n", X1)
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
```

```python

# Create a graph object
MyGraph = nx.Graph()

# Add nodes
# Green Line
MyGraph.add_node('A', npos=(33, 28), ccn='#0000FF')
MyGraph.add_node('B', npos=(61, 56), ccn='#0000FF')
MyGraph.add_node('C', npos=(35, 81), ccn='#0000FF')
MyGraph.add_node('D', npos=(45, 37), ccn='#0000FF')
# Orange Line
MyGraph.add_node('E', npos=(32, 99), ccn='#FF4500')
MyGraph.add_node('F', npos=(56, 23), ccn='#FF4500')
MyGraph.add_node('G', npos=(91, 85), ccn='#FF4500')
MyGraph.add_node('H', npos=(24, 36), ccn='#FF4500')
# Blue Line
MyGraph.add_node('I', npos=(99, 70), ccn='#00FF00')
MyGraph.add_node('J', npos=(82, 73), ccn='#00FF00')
MyGraph.add_node('K', npos=(81, 72), ccn='#00FF00')
MyGraph.add_node('L', npos=(47, 38), ccn='#00FF00')

# Connect nodes
# Green Line
MyGraph.add_edge('A', 'B', cce='#0000FF')
MyGraph.add_edge('B', 'C', cce='#0000FF')
MyGraph.add_edge('C', 'D', cce='#0000FF')
# Orange Line
MyGraph.add_edge('E', 'F', cce='#FF4500')
MyGraph.add_edge('F', 'G', cce='#FF4500')
MyGraph.add_edge('G', 'H', cce='#FF4500')
# Blue Line
MyGraph.add_edge('I', 'J', cce='#00FF00')
MyGraph.add_edge('J', 'K', cce='#00FF00')
MyGraph.add_edge('K', 'L', cce='#00FF00')

# Extract attributes from the graph to dictionaries
pos = nx.get_node_attributes(MyGraph, 'npos')
nodecolour = nx.get_node_attributes(MyGraph, 'ccn')
edgecolour = nx.get_edge_attributes(MyGraph, 'cce')

# Place the dictionary values in lists
NodeList = list(nodecolour.values())
EdgeList = list(edgecolour.values())

# Set the size of the figure
plt.figure(figsize=(10, 7))

# Display the names of the stations
plt.text(35, 30, s='Heron Quays', rotation=0)
plt.text(38, 81, s='South Quays', rotation=0)
plt.text(34, 99, s='Mudchute', rotation=0)
plt.text(63, 56, s='Island gardens', rotation=0)

plt.text(93, 85, s='Canary Wharf', rotation=0)
plt.text(20, 40, s='canada water', rotation=0)
plt.text(50, 38, s='Bermondsey', rotation=0)

plt.text(35, 42, s='Wembley Park', rotation=0)
plt.text(58, 23, s='neasden', rotation=0)
plt.text(100, 70, s='Dollishill', rotation=0)

# Draw the nodes and the edges
nx.draw_networkx(MyGraph, pos, node_color=NodeList)
nx.draw_networkx_edges(MyGraph, pos, edge_color=EdgeList)

# Visualise the graph
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def DrawBox(x, y, size, r, g, b):
    if r < 0:
        r = int(0)
    if g < 0:
        g = int(0)
    if b < 0:
        b = int(0)
    if r > 255:
        r = int(255)
    if g > 255:
        g = int(255)
    if b > 255:
        b = int(255)
    for i in range(0, int(size)):
        plt.plot([x, x + size+10], [y + i, y + i], '#{:02x}{:02x}{:02x}'.format(r, g, b))


#assigning a dataset to dataframe using pandas library
df = pd.read_csv("C:/Users/POORVAA/Desktop/HeatMap.csv")
# Print the content on the screen
print(df.head(7))


plt.axis([0, 600, 0, 400])
plt.xticks([])
plt.yticks([])

Min = int(min(df.min(numeric_only=True)))
Max = int(max(df.max(numeric_only=True)))

BoxSize = int(40)
OffsetX = int(15)
OffsetY = int(12)

# Generating the heat map
for i in range(0, df.shape[0]):
    for j in range(1, df.shape[1]):
        ColourCode = int(((df.values[i, j]-Min)/(Max-Min))*255)
        DrawBox(20+BoxSize*j, 300-BoxSize*i, BoxSize, ColourCode, 0, 0)
        plt.text(OffsetX+20+BoxSize*j, OffsetY+300-BoxSize*i, str(df.values[i, j]), color='white')

for i in range(0, 256):
    plt.plot([560, 580], [i + 60, i + 60], '#{:02x}{:02x}{:02x}'.format(int(i), 0, 0))
plt.text(585, 58, Min)
plt.text(585, 312, Max)

plt.text(72, 20, 'Jan')
plt.text(112, 20, 'Feb')
plt.text(152, 20, 'Mar')
plt.text(192, 20, 'Apr')
plt.text(232, 20, 'May')
plt.text(272, 20, 'Jun')
plt.text(312, 20, 'Jul')
plt.text(352, 20, 'Aug')
plt.text(392, 20, 'Sep')
plt.text(432, 20, 'Oct')
plt.text(472, 20, 'Nov')
plt.text(512, 20, 'Dec')

plt.text(5, 315, str(df.values[0, 0]))
plt.text(5, 275, str(df.values[1, 0]))
plt.text(5, 235, str(df.values[2, 0]))
plt.text(5, 195, str(df.values[3, 0]))
plt.text(5, 155, str(df.values[4, 0]))
plt.text(5, 115, str(df.values[5, 0]))
plt.text(5, 75, str(df.values[6, 0]))

plt.show()
```
